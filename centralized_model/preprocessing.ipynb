```python
import pandas as pd

# loading dataset
df = pd.read_csv('../data/heart.csv')

# Basic operations
print(df.shape)
print(df.columns)
print(df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Encoding variables using one-hot encoding (Making them 0/1)
df_encoded = pd.get_dummies(df, columns=['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], drop_first=True)

# Splitting features and target
X = df_encoded.drop('HeartDisease', axis=1)
y = df_encoded['HeartDisease']

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Splitting the Training and Testing data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Checking shape of both
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
import numpy as np

# Saving preprocessed arrays to numpy files
np.save('../data/X_train.npy', X_train)
np.save('../data/X_test.npy', X_test)
np.save('../data/y_train.npy', y_train)
np.save('../data/y_test.npy', y_test)
```
